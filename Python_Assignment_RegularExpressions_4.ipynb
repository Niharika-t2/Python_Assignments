#**1.Select all email addresses.**

```python
#reading data from the text file 

fp = open(r"data.txt", "r").read()
print(fp)
```

```python
#printing all email addresses

import re

match = re.findall("\S+@\S+",fp)
print(match)  
```

#**2.Select all addresses with pincode starting with 2 or 1.**

```python
import re 

match=re.findall("[1|2]\d{4}",fp)
print(match)
```

#**3.Select all mobile numbers starting with 800**

```python
import re

match = re.findall("[8][0][0]-\d{3}-\d{4}",fp)
print(match)
```

#**4.Write a python program to find the occurence and position of the substrings within a string.**

```python
import re

substring = 'Wilson'
match = [(string, string.start()) for string in re.finditer(substring, fp)]

print(match)
```

#**5.Write a python program to match a string that contains only upper and lowercase letters, numbers and underscores.**

```python
import re

pattern = '[A-Za-z]*|[0-9_]*'
match = re.findall(pattern,fp)

print(match)
```
